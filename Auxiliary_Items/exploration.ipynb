```python
import numpy as np
import pickle
import os
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
color = [["red"], ["green"], ["blue"], ["orange"], ["purple"]]
idx = 0
d = {}
for file in os.listdir("sim"):
    file = "_".join(file.split(".")[:-1])
    scores = pickle.load(open("sim/" + file + ".npy", "rb"))
    if "euclidean" in file:
        scores /= np.max(scores)
    d["_".join(file.split("_")[2:4])] = np.sort(scores)[::-1]

d = pd.DataFrame(d)
columns = [col for col in d.columns if "euclidean" in col or "Jaccard" in col]

print(columns)
sns.scatterplot(data = d[columns])
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
columns = [col for col in d.columns if "cos" in col or "Jaccard" in col]

print(columns)
sns.scatterplot(data = d[columns])
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
